#### Importing Necessary Libraries & "Stocks" DataFrame

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
stocks = pd.read_csv('stocks.csv')  # Compiled in Notebook 3
```

#### Importing Stocks DataFrame, Long Portfolio, & Short Portfolio 

```python
stocks = pd.read_csv('stocks.csv')  # Read in csv exported from Notebook 3
long_portfolio = pd.read_csv('long_portfolio.csv') # Read in csv exported from Notebook 2
short_portfolio = pd.read_csv('short_portfolio.csv') # Read in csv exported from Notebook 2
```

#### Ensure Stocks DataFrame only includes Long & Short Portfolio tickers
_20 Most Frequent from each_

```python
long_portfolio_list = long_portfolio['Tickers'].value_counts().head(20).index.tolist()
short_portfolio_list = short_portfolio['Tickers'].value_counts().head(20).index.tolist()

all_stocks = long_portfolio_list + short_portfolio_list

# Subsetting Stocks on list of most frequent stock tickers 
stocks = stocks[stocks['Tickers'].isin(all_stocks)]

# Grab copy to use as finalized Stocks DataFrame 
final_stocks = stocks.copy()
```

### Datatype Conversion
---
`Tickers` : Stock each observation/date is referencing 
- `Categorical` data
- No need for datatype conversion

`Close` : Final stock price for each observation/date
- Numerical data can not be utilized as an object
- Convert to `float` 

`Time` :  Timestamp for each observation/date 
- As with Ideas, convert to a `datetimeobject` with Pandas
- A step further:
  -   Pandas datetime objects can access different parts of the date
  -   We will be analyzing stock data on a weekly basis
  -   Extract week number 

#### Converting 'Close' to float.. 
- Simply converting with `'astype'` threw an error because nulls are recorded as text
   `'null'`
- Dropping these indexes 

```python
stocks = stocks.loc[stocks['Close'] != 'null'] # Removing indexes  

stocks['Close'] = stocks['Close'].astype(float) # Converting to float 
```

#### Converting 'Date' to datetime object.. 

```python
stocks['Date'] = pd.to_datetime(stocks['Date']) # Converting to datetime 
```

```python
stocks['Week'] = stocks['Date'].dt.week
stocks['Month'] = stocks['Date'].dt.month
```

## Visualizing Feature Distributions & Frequenices
---
### What is the distribution for closing prices?

```python
plt.figure(figsize=(46,30))

sns.violinplot(stocks['Close'],palette='inferno',width=1)
plt.xlim(0,4000)
plt.xlabel('Closing Price($)',fontsize=60)
plt.ylabel('Frequency',fontsize=60,rotation=0,labelpad=145)
plt.title('Distribution- Stock Prices',fontsize=90)
plt.xticks(fontsize=50)
plt.yticks(fontsize=22)
```

```python
sns.set_style('whitegrid')
sns.despine()

d = stocks['Close']

plt.figure(figsize=(16,10))
sns.distplot(d, bins=2500,color='m')

plt.xlabel('Closing Price($)',fontsize=32)
plt.ylabel('Percentage',fontsize=32,rotation=0,labelpad=90)
plt.xlim(0,200)
plt.title('Distribution- Stock Prices',fontsize=50)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
```

### What stocks have the highest closing prices?

```python
# Bar graph to visualize monthly articles

avg_close = stocks.groupby('Tickers')['Close'].mean()

plt.figure(figsize=(16,10))
sns.barplot(x=avg_close.index,
            y=avg_close.values,
            palette='inferno')


plt.xlabel('Ticker',fontsize=32)
plt.ylabel('Closing Price($)',fontsize=32,rotation=0,labelpad=110)
plt.title('Avg. Closing Prices by Ticker',fontsize=50)
plt.xticks(fontsize=22,rotation=90)
plt.yticks(fontsize=22)
```

### Are closing prices higher certain times of the year?

```python
monthly_close = stocks.groupby('Month')['Close'].mean()

# Bar graph to visualize monthly articles

plt.figure(figsize=(16,10))
sns.barplot(x=monthly_close.index,
            y=monthly_close.values,
            palette='inferno')


plt.xlabel('Month',fontsize=32)
plt.ylabel('Closing Price($)',fontsize=32,rotation=0,labelpad=130)
plt.title('Avg. Closing Prices by Month',fontsize=50)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
```

## Exploring Long and Short Portfolio Performance
---
#### Over the past year, how have Long and Short Portfolio stocks performed overall?

```python
# Subsetting Stocks to only include:

# Long Portfolio tickers
long = stocks[stocks['Tickers'].isin(long_portfolio_list)].reset_index(drop=True)

# Short Portfolio tickers
short = stocks[stocks['Tickers'].isin(short_portfolio_list)].reset_index(drop=True)
```

```python
# Transforming Stock subsets into pivot tables- indexing on Week # and ticker

# Long Portfolio pivot 
long_pivot = long.pivot_table('Close', index='Week', columns='Tickers')

# Short Portfolio pivot 
short_pivot = short.pivot_table('Close', index='Week', columns='Tickers')
```

```python
long_pivot.head(3)  # Inspecting head
```

```python
# Plot the results
fig, ax = plt.subplots(figsize=(10,8))
long_pivot.plot(ax=ax, 
                 ylim=(0,440), yticks=(np.arange(0,440,step=25)),
                 xlim=(0,52),xticks=(np.arange(0,52,step=5)));
plt.xlabel('Week',fontsize= 15)
plt.xticks(fontsize=12.5)
plt.ylabel('Closing Price',fontsize=15)
plt.yticks(fontsize=12.5)
plt.title('Long Idea Stocks Over Past Year',fontsize=25)
plt.legend(loc='upper left')
```

```python
short_pivot.head(3)  # Inspecting head
```

```python
# Plot the results
fig, ax = plt.subplots(figsize=(10,8))
short_pivot.plot(ax=ax, 
                 ylim=(0,500), yticks=(np.arange(0,450,step=25)),
                 xlim=(0,52),xticks=(np.arange(0,52,step=5)));
plt.xlabel('Week',fontsize= 15)
plt.xticks(fontsize=12.5)
plt.ylabel('Closing Price',fontsize=15)
plt.yticks(fontsize=12.5)
plt.title('Short Idea Stocks Over Past Year',fontsize=25)
plt.legend(loc='upper left')
```
